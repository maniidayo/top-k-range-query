```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output books.csv', names=['k','BruteForce','BruteForceX','WIT','WITX','WIT+Bucket','WIT+BucketX'], skiprows=1)
plt.figure(figsize=(6, 4))
plt.plot(df['k'], df['BruteForceX'], marker='o', label='BruteForce')
plt.plot(df['k'], df['WITX'], marker='o', label='WIT')
plt.plot(df['k'], df['WIT+BucketX'], marker='o', label='WIT+Bucket')
plt.xlabel('k')
plt.ylabel('Time (ms)')
plt.title('books')
plt.legend()
plt.grid(True)
plt.show()
```
